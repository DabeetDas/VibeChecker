```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_csv('merged_clean_data.csv')

plt.scatter(df['score'],df['num_comments'])
plt.title("num_comments vs score(cleaned)")
plt.xlabel("score")
plt.ylabel("num_comments")
plt.show()

```

```python
df1=pd.read_csv("Merged_data.csv")
x=0
y=0
labelS=['image','no image']
for index,rows in df1.iterrows():
  crct_extensions=('.jpg', '.jpeg', '.png', '.gif', '.webp')
  url_val=rows['image_url']
  if not pd.isna(url_val) and str(url_val).lower().endswith(crct_extensions) :
    x+=1
  else:
    y+=1
print(x,y)
plt.title("url invalid or not present vs crct url(uncleaned)")
plt.pie([x,y],labels=labelS)
plt.show()
```
